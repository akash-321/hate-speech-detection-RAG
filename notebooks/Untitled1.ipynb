```python
religion_cols = [
    'annotator_religion_atheist',
    'annotator_religion_buddhist',
    'annotator_religion_christian',
    'annotator_religion_hindu',
    'annotator_religion_jewish',
    'annotator_religion_mormon',
    'annotator_religion_muslim',
    'annotator_religion_nothing',
    'annotator_religion_other'
]

religion_counts = (
    df[religion_cols]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

religion_counts.columns = ["Religion", "Number of annotations"]
religion_counts
```

```python
target_cols = [
    'target_religion_atheist',
 'target_religion_buddhist',
 'target_religion_christian',
 'target_religion_hindu',
 'target_religion_jewish',
 'target_religion_mormon',
 'target_religion_muslim',
 'target_religion_other',
 'target_religion',
 'target_origin_immigrant',
 'target_origin_migrant_worker',
 'target_origin_specific_country',
 'target_origin_undocumented',
 'target_origin_other',
 'target_origin',
 'target_gender_men',
 'target_gender_non_binary',
 'target_gender_transgender_men',
 'target_gender_transgender_unspecified',
 'target_gender_transgender_women',
 'target_gender_women',
 'target_gender_other',
 'target_gender',
 'target_sexuality_bisexual',
 'target_sexuality_gay',
 'target_sexuality_lesbian',
 'target_sexuality_straight',
 'target_sexuality_other',
 'target_sexuality',
 'target_age_children',
 'target_age_teenagers',
 'target_age_young_adults',
 'target_age_middle_aged',
 'target_age_seniors',
 'target_age_other',
 'target_age',
 'target_disability_physical',
 'target_disability_cognitive',
 'target_disability_neurological',
 'target_disability_visually_impaired',
 'target_disability_hearing_impaired',
 'target_disability_unspecific',
 'target_disability_other',
 'target_disability',
 'target_politics_alt_right',
 'target_politics_communist',
 'target_politics_conservative',
 'target_politics_democrat',
 'target_politics_green_party',
 'target_politics_leftist',
 'target_politics_liberal',
 'target_politics_libertarian',
 'target_politics_republican',
 'target_politics_socialist',
 'target_politics_other',
 'target_politics',
]

religion_target_table = {}

for rel in religion_cols:
    rel_name = rel.replace("annotator_religion_", "")
    subset = df[df[rel] == 1]
    religion_target_table[rel_name] = subset[target_cols].sum()

religion_target_df = pd.DataFrame(religion_target_table).T
religion_target_df
```

```python
ordinal_cols = [
    'sentiment',
    'respect',
    'insult',
    'humiliate',
    'violence',
    'genocide',
    'hatespeech'
]

religion_ordinal_table = {}

for rel in religion_cols:
    rel_name = rel.replace("annotator_religion_", "")
    subset = df[df[rel] == 1]
    religion_ordinal_table[rel_name] = subset[ordinal_cols].mean()

religion_ordinal_df = pd.DataFrame(religion_ordinal_table).T
religion_ordinal_df
```

```python
binary_hate_by_religion = {}

for rel in religion_cols:
    rel_name = rel.replace("annotator_religion_", "")
    subset = df[df[rel] == 1]
    binary_hate_by_religion[rel_name] = subset["hatespeech"].mean()

binary_hate_df = (
    pd.Series(binary_hate_by_religion)
    .reset_index()
)

binary_hate_df.columns = ["Religion", "Proportion hate speech"]
binary_hate_df
```

```python
df["binary_hate"] = (df["hatespeech"] > 0).astype(int)
```

```python
df[["hatespeech", "binary_hate"]].head()
```

```python
disagreement_by_religion = {}

for rel in religion_cols:
    rel_name = rel.replace("annotator_religion_", "")
    subset = df[df[rel] == 1]
    disagreement_by_religion[rel_name] = (
        subset.groupby("comment_id")["binary_hate"].std().mean()
    )

pd.Series(disagreement_by_religion)
```

```python
def score_to_label(score):
    if score > 0.5:
        return "Hate speech"
    elif score < -1:
        return "Counter"
    else:
        return "Neutral"

df["score_label"] = df["hate_speech_score"].apply(score_to_label)
```

```python
label_distribution = (
    df["score_label"]
    .value_counts(normalize=True)
    .reset_index()
)

label_distribution.columns = ["Category", "Proportion of annotations"]
label_distribution
```

```python
comparison_table = pd.crosstab(
    df["hatespeech"],
    df["score_label"],
    normalize="index"
)

comparison_table
```
